```python
#Welcome to learn the concepts of the Matplotlib
#This is a very Useful library to plot the Graphs
import matplotlib
print(matplotlib.__version__)
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Learning to Draw linear plots
xpoints=np.array([0,6])
ypoints=np.array([0,250])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
# Understanding the .plot function
'''This function takes in 2 Argumenst x-cor and y-cor and byDef creates linear lines'''
xpoints=np.array([1,8])
ypoints=np.array([3,10])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
# Plotting withOut line
xpoints=np.array([1,8])
ypoints=np.array([3,10])

plt.plot(xpoints,ypoints,'o')# This 'o' specifies rings without lines
plt.show
```

```python
# We can plot as many points as we want , just make sure that  the number of x-cor is equal
# to the number of y-cor
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
# Default values of the x-cor
# is set to [0,1,2,3,...]
ypoints=np.array([3,8,1,10,5,7])
plt.plot(ypoints)
plt.show()
```

```python
# Understanding the keyWord 'Marker'
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])

plt.plot(xpoints,ypoints,marker='*')
plt.show()


plt.plot(xpoints,ypoints,marker='P')
plt.show()

plt.plot(xpoints,ypoints,marker='D')
plt.show()
```

```python
# Format String fmt
# This is used to define marker|line|color

ypoints=np.array([0,10,20,30,40])
plt.plot(ypoints,'o:r')
plt.show()
```

```python
# Understanding the marker size
plt.plot(ypoints,marker='P',ms=20)
plt.show()
```

```python
# Similarly  one can use markerEdge color
plt.plot(ypoints,marker='o',ms=5,mec='r')
plt.show()
```

```python
# Similarly markerfacecolor with keyword mfc
plt.plot(ypoints,'o:m',ms=5,mec='r',mfc='g')
plt.show()
```

```python
# Welcome the concepts of the Labels and Title
import matplotlib.pyplot as plt
import numpy as np

x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y)


# Add the labels here, after plt.plot()

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")



plt.show()
```

```python
# create a title for plot
import matplotlib.pyplot as plt
import numpy as np

x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y)


# Add the labels here, after plt.plot()
plt.title('Sports Watch Data')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")



plt.show()
```

```python
# we can also set font properties for the xlabel, ylabel and title usin Dictionary

font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}

plt.title('Sports Watch Data',fontdict=font1)
plt.xlabel("Average Pulse",fontdict=font2)
plt.ylabel("Calorie Burnage",fontdict=font2)

plt.plot(x,y)

plt.show()
```

```python
# Using loc attribute to position to the  title
plt.title('Sports Watch Data',loc='left')
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')
plt.plot(x,y)
plt.show()
```

```python
# Adding Grid Lines to a Plot
import matplotlib.pyplot as plt
import numpy as np

x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.title('Sports Watch Data')
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')
plt.plot(x,y)

# Just add plt.grid befor plt.show() function
plt.grid()
plt.show()
```

```python
# We can also use the axis attribute in the grid function , legal values are x,y,both
plt.plot(x,y)
plt.grid(axis='both',ls='dotted',c='g',linewidth='3')
plt.show()
```

```python
# Welcome to the Concepts of the Subplots
'''With subplot() function you can draw multiple plots in one figure'''

import matplotlib.pyplot as plt
import numpy as np

#plot 1 : Desc
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])


# undestandint what does this line mean
# This line says that we will have 1 rows and 2 plots and this is the plot number1
plt.subplot(1,2,1)
plt.plot(x,y)

#plot 2 : Desc
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()


# As you can see in the result we have 2 fig in 1 rows
```

```python
# Displaying 2 fig in 2 rows i.e 2 rows and cols


#plot1
x=np.array([1,2,3,4])
y=np.array([5,6,7,8])

plt.subplot(2,1,1)
plt.plot(x,y)

#plot2
x=np.array([9,10,11,12])
y=np.array([13,14,15,16])

plt.subplot(2,1,2)
plt.plot(x,y)

# Now show the Entire plot done so far
plt.show()
```

```python
# We can give title to each figure and superTitle as well
# Displaying 2 fig in 2 rows i.e 2 rows and cols


#plot1
x=np.array([1,2,3,4])
y=np.array([5,6,7,8])

plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('First Plot')
plt.xlabel('X1-axis')
plt.ylabel('Y1-axis')

#plot2
x=np.array([9,10,11,12])
y=np.array([13,14,15,16])

plt.subplot(2,1,2)
plt.plot(x,y)
plt.title('Second Plot')
plt.xlabel('X2-axis')
plt.ylabel('Y2-axis')

# Now show the Entire plot done so far
plt.suptitle('My Sporst watch Data')
plt.show()
```
